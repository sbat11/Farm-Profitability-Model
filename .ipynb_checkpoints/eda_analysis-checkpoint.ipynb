```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
incomeDF = pd.read_excel("cleanIncomeData.xlsx", sheet_name="Cleaned Data")
print(incomeDF.head())
```

```python
expenseDF = pd.read_excel("cleanProductionExpensesData.xlsx", sheet_name="Cleaned Data")
print(expenseDF.head())
```

```python
mergedDF = pd.concat([incomeDF, expenseDF], ignore_index=True)
print(mergedDF)
```

```python
# Convert numeric columns to numeric data types (ensuring plots work correctly)
numeric_columns = mergedDF.columns[1:]  # Exclude the first column (category names)
mergedDF[numeric_columns] = mergedDF[numeric_columns].apply(pd.to_numeric, errors='coerce')
print(mergedDF[numeric_columns])
```

```python
# --- HISTOGRAM OF EXPENSE AND INCOME DISTRIBUTIONS ---
plt.figure(figsize=(12, 6))
plt.hist(mergedDF[numeric_columns].T, bins=10, edgecolor='black', alpha=0.6)
plt.title("Distribution of Farm Expenses and Gross Cash Income")
plt.xlabel("Value Range")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
# --- TIME-SERIES TREND PLOT ---
plt.figure(figsize=(12, 6))
years = numeric_columns.astype(str)  # Ensure years are string for plotting

# Select key expenses and plot their trends
selected_expenses = ['cash expenses > excluding net rent to landlords', 'intermediate product expenses > seed', 'intermediate product expenses > fuel and oils', 'intermediate product expenses > fertilizer, lime, and soil conditioner', 'labor expenses > contract labor']
for category in selected_expenses:
    values = mergedDF[mergedDF["cat"] == category].iloc[:, 1:].values.flatten()
    if values.size > 0:
        plt.plot(years, values, label=category, linestyle="--")
```

```python
# Plot Gross Cash Income
gci_values = mergedDF[mergedDF["cat"] == "gross cash income"].iloc[:, 1:].values.flatten()
if gci_values.size > 0:
    plt.plot(years, gci_values, label="Gross Cash Income", linewidth=2, color="black")

plt.legend()
plt.title("Time-Series Trends of Farm Expenses and Gross Cash Income")
plt.xlabel("Year")
plt.ylabel("Amount ($100 Billion)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# --- CORRELATION HEATMAPS ---
selected_categories = [
    "gross cash income",
    "labor expenses > contract labor",
    "property taxes and fees",
    "real estate taxes",
    "intermediate product expenses > electricity",
    "intermediate product expenses > pesticide",
]

selected_categories2 = [
    "gross cash income",
    "intermediate product expenses > feed",
    "intermediate product expenses > seed",
    "labor expenses > cash labor",
    "labor expenses > contract labor",
    "intermediate product expenses > fertilizer, lime, and soil conditioner",
]

for idx, (title, categories) in enumerate([
    ("Correlation Heatmap (Selected Categories)", selected_categories),
    ("Correlation Heatmap2 (Selected Categories)", selected_categories2)
]):
    plt.figure(figsize=(10, 8))
    filtered_df = mergedDF[mergedDF['cat'].isin(categories)].set_index("cat").T
    filtered_df = filtered_df.apply(pd.to_numeric)
    corr_matrix = filtered_df.corr()
    sns.heatmap(corr_matrix, cmap="coolwarm", annot=True, linewidths=1.0)
    plt.title(title)
    plt.xticks(rotation=45, ha="right")
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()

    if "gross cash income" in corr_matrix.columns:
        target_series = corr_matrix['gross cash income'].drop('gross cash income')
        top_features = target_series.sort_values(ascending=False).head(5)
        print(f"Top correlated features with gross cash income in '{title}':\n{top_features}\n")
```

CONCLUSION: 
    1. The top correlated features with gross cash income were:
     - property taxes and fees                        0.941402
     - intermediate product expenses > pesticide      0.902076
     - intermediate product expenses > electricity    0.872844
     - intermediate product expenses > fertilizer, lime, and soil conditioner    0.913032
     - intermediate product expenses > seed                                      0.901391
     - labor expenses > cash labor                                               0.879840

Why these features: 
         - These are independent variables, not subcomponents of gross income.
         - They're real operational costs that contribute to productivity.
         - They cover different dimensions: land use (taxes), crop production (seed, fertilizer, pesticide), utilities (electricity), and labor.
